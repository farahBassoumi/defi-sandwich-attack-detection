```python
import pandas as pd
import numpy as np

# 1. Load Dataset (in your case, from a merged df directly)
def load_dataset() -> pd.DataFrame:
    print("📥 Loading merged dataset...")
    return merged.copy()

def rename_pending_time_column(df):
    """
    Rename the 'time_pending' column to 'time_pending_by_blocknative' if it exists.
    """
    if 'time_pending' in df.columns:
        df = df.rename(columns={'time_pending': 'time_pending_by_blocknative'})
    return df

# 2. Drop unwanted columns
def drop_unwanted_columns(df: pd.DataFrame) -> pd.DataFrame:
    print("🧹 Dropping unwanted columns...")
    cols_to_drop = [
        'Unnamed: 0',
        *[col for col in df.columns if len(col) > 1000 or ',' in col]
    ]
    return df.drop(columns=cols_to_drop, errors='ignore')

# 3. Report and drop duplicates
def report_and_drop_duplicates(df: pd.DataFrame) -> pd.DataFrame:
    dup_percent = df.duplicated().mean() * 100
    print(f"📊 Duplicate rows: {dup_percent:.2f}%")
    return df.drop_duplicates()

# 4. Clean column names
def clean_column_names(df: pd.DataFrame) -> pd.DataFrame:
    print("🧼 Cleaning column names...")
    df = rename_pending_time_column(df)
    df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]
    return df

# 5. Optimize numeric memory usage
def optimize_dataframe(df: pd.DataFrame) -> pd.DataFrame:
    print("🔧 Optimizing memory usage...")

    for col in df.select_dtypes(include=['int', 'float']).columns:
        if set(df[col].dropna().unique()).issubset({0, 1}):
            df[col] = df[col].astype(bool)
            print(f"🔁 Converted '{col}' to bool")
        else:
            dtype_str = str(df[col].dtype)
            if 'int' in dtype_str:
                df[col] = pd.to_numeric(df[col], downcast='unsigned')
                print(f"📉 Downcasted '{col}' to {df[col].dtype}")
            elif 'float' in dtype_str:
                df[col] = pd.to_numeric(df[col], downcast='float')
                print(f"📉 Downcasted '{col}' to {df[col].dtype}")

    return df


# 6. Memory usage reporting
def report_memory(df: pd.DataFrame, label="") -> None:
    mem = df.memory_usage(deep=True).sum() / 1024**2
    print(f"💾 {label} Memory usage: {mem:.2f} MB")

# 7. Show null value insights
def show_null_insights(df: pd.DataFrame) -> None:
    nulls = df.isnull().mean() * 100
    nulls = nulls[nulls > 0].sort_values(ascending=False)
    print("📉 Columns with missing values:")
    print(nulls.to_string())

# 8. Show full head
def show_full_head(df: pd.DataFrame, rows=5) -> pd.DataFrame:
    with pd.option_context('display.max_columns', None):
        return df.head(rows)


def show_value_percentages(df: pd.DataFrame, top_n: int = 5) -> None:
    print("\n📊 Percentage of each value (top {top_n}) per column:\n")

    for col in df.columns:
        print(f"🔹 Column: {col}")

        try:
            value_counts = df[col].value_counts(normalize=True, dropna=False) * 100
            print(value_counts.head(top_n).round(2).to_string())
            print("-" * 40)
        except Exception as e:
            print(f"⚠️ Could not process column {col}: {e}")


def update_pending_times_and_clean(df: pd.DataFrame) -> pd.DataFrame:
    print("🔁 Updating pending times and removing raw pending rows...")

    # Make sure detecttime is datetime
    df['detecttime'] = pd.to_datetime(df['detecttime'], errors='coerce')

    # Split pending and confirmed sets
    pending_df = df[df['status'] == 'pending']
    confirmed_df = df[df['status'] == 'confirmed']

    # Keep only hashes that are present in both
    common_hashes = set(pending_df['hash']).intersection(set(confirmed_df['hash']))

    # Only process those
    pending_df = pending_df[pending_df['hash'].isin(common_hashes)]
    confirmed_df = confirmed_df[confirmed_df['hash'].isin(common_hashes)]

    # Map pending times to confirmed df
    pending_times = pending_df[['hash', 'detecttime']].rename(columns={'detecttime': 'pending_detecttime'})
    confirmed_df = confirmed_df.merge(pending_times, on='hash', how='left')

    # Compute time diff in seconds
    confirmed_df['computed_time_pending'] = (
        confirmed_df['detecttime'] - confirmed_df['pending_detecttime']
    ).dt.total_seconds()

    # Update time_pending column
    confirmed_df['time_pending'] = confirmed_df.apply(
        lambda row: row['computed_time_pending']
        if pd.isna(row['time_pending']) else max(row['time_pending'], row['computed_time_pending']),
        axis=1
    )

    # Drop helper columns
    confirmed_df = confirmed_df.drop(columns=['pending_detecttime', 'computed_time_pending'])

    # Remove original pending rows
    df_cleaned = df[df['status'] != 'pending']

    # Replace old confirmed rows with updated ones
    df_cleaned = df_cleaned[~df_cleaned['hash'].isin(common_hashes)]
    df_cleaned = pd.concat([df_cleaned, confirmed_df], ignore_index=True)

    print("✅ time_pending updated and pending rows removed.")
    return df_cleaned



# 9. Run entire pipeline
def run_pipeline() -> pd.DataFrame:
    df = load_dataset()
    report_memory(df, "Original")
    df = drop_unwanted_columns(df)
    df = rename_pending_time_column(df)
    df = report_and_drop_duplicates(df)
    df = clean_column_names(df)
    show_null_insights(df)  # <--- Only shows insights, doesn't drop
    df = optimize_dataframe(df)
    df = update_pending_times_and_clean(df)

    show_value_percentages(df.copy(),10)

    report_memory(df, "Optimized")

    print("✅ Pipeline complete.")
    return df
```

```python
cleaned_df=run_pipeline()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

class InteractiveSybilDetectionPipeline:
    def __init__(self, df):
        """
        Initialize the pipeline with the raw dataset
        """
        self.df = df.copy()
        self.processed_df = None
        self.feature_importance = None
        self.scaler = StandardScaler()
        self.label_encoders = {}
        self.step_history = []

        print("🚀 Interactive Sybil Detection Pipeline Initialized")
        print("=" * 60)
        self._show_dataset_overview(self.df, "Original Dataset")

    def _show_dataset_overview(self, df, title):
        """Show comprehensive dataset overview"""
        print(f"\n📊 {title} Overview:")
        print("-" * 40)
        print(f"Shape: {df.shape}")
        print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

        if 'is_sybil' in df.columns:
            target_dist = df['is_sybil'].value_counts()
            print(f"Target distribution:")
            for val, count in target_dist.items():
                pct = count / len(df) * 100
                print(f"  {val}: {count} ({pct:.2f}%)")

        print(f"\nData types:")
        print(df.dtypes.value_counts())

        print(f"\nMissing values:")
        missing = df.isnull().sum()
        missing_pct = (missing / len(df) * 100).round(2)
        missing_summary = pd.DataFrame({
            'Missing Count': missing,
            'Missing %': missing_pct
        })
        print(missing_summary[missing_summary['Missing Count'] > 0].head(10))

        print(f"\nFirst 3 rows:")
        print(df.head(3))

    def preview_aggregation(self):
        """
        Preview what the aggregation will create without actually doing it
        """
        print("\n🔍 PREVIEW: Feature Aggregation Analysis")
        print("=" * 60)

        # Show unique addresses
        unique_addresses = self.df['fromaddress'].nunique()
        total_transactions = len(self.df)
        avg_tx_per_address = total_transactions / unique_addresses

        print(f"📈 Aggregation Summary:")
        print(f"  • Total transactions: {total_transactions:,}")
        print(f"  • Unique addresses: {unique_addresses:,}")
        print(f"  • Average transactions per address: {avg_tx_per_address:.2f}")

        # Show transaction distribution per address
        tx_per_address = self.df['fromaddress'].value_counts()
        print(f"\n📊 Transaction Distribution per Address:")
        print(f"  • Min transactions: {tx_per_address.min()}")
        print(f"  • Max transactions: {tx_per_address.max()}")
        print(f"  • Median transactions: {tx_per_address.median()}")
        print(f"  • Mean transactions: {tx_per_address.mean():.2f}")

        # Show top addresses by transaction count
        print(f"\n🏆 Top 10 Most Active Addresses:")
        top_addresses = tx_per_address.head(10)
        for addr, count in top_addresses.items():
            sybil_status = self.df[self.df['fromaddress'] == addr]['is_sybil'].any()
            status_emoji = "⚠️" if sybil_status else "✅"
            print(f"  {status_emoji} {addr[:10]}...{addr[-8:]}: {count} transactions")

        # Show what features will be created
        numeric_cols = ['nonce', 'gas', 'gasprice', 'value', 'blockspending',
                       'time_pending_by_blocknative', 'gasused', 'maxpriorityfeepergas',
                       'maxfeepergas', 'basefeepergas', 'time_pending', 'was_evicted',
                       'was_rejected']

        available_numeric = [col for col in numeric_cols if col in self.df.columns]

        print(f"\n🎯 Features to be Created:")
        print(f"  • Statistical features for {len(available_numeric)} numeric columns:")
        for col in available_numeric:
            print(f"    - {col}: count, mean, std, min, max, median")

        print(f"  • Behavioral features:")
        behavioral_features = [
            "total_transactions", "successful_transactions", "failed_transactions",
            "unique_recipients", "primary_tx_type", "primary_region",
            "dropped_transactions", "activity_timespan_seconds", "active_days",
            "success_rate", "failure_rate", "unique_recipient_ratio",
            "drop_rate", "tx_frequency"
        ]
        for feature in behavioral_features:
            print(f"    - {feature}")

        print(f"\n📏 Expected output shape: ({unique_addresses}, ~{len(available_numeric)*6 + len(behavioral_features)})")

        return input("\n❓ Proceed with aggregation? (y/n): ").lower().strip() == 'y'

    def aggregate_features(self, confirm=True):
        """
        Create aggregated features grouped by tx_from address (fromaddress)
        """
        if confirm and not self.preview_aggregation():
            print("❌ Aggregation cancelled by user")
            return None

        print("\n🔄 Executing Feature Aggregation...")

        # Numeric aggregations
        numeric_cols = ['nonce', 'gas', 'gasprice', 'value', 'blockspending',
                       'time_pending_by_blocknative', 'gasused', 'maxpriorityfeepergas',
                       'maxfeepergas', 'basefeepergas', 'time_pending', 'was_evicted',
                       'was_rejected']

        aggregated_features = []

        # Basic statistics for numeric columns
        for col in numeric_cols:
            if col in self.df.columns:
                agg_data = self.df.groupby('fromaddress')[col].agg([
                    'count', 'mean', 'std', 'min', 'max', 'median'
                ]).add_prefix(f'{col}_')
                aggregated_features.append(agg_data)

        # Transaction behavior patterns
        tx_patterns = self.df.groupby('fromaddress').agg({
            'hash': 'count',
            'status': lambda x: (x == 'confirmed').sum(),
            'failurereason': lambda x: (x != 'none').sum(),
            'toaddress': 'nunique',
            'type': lambda x: x.mode().iloc[0] if len(x) > 0 else 0,
            'region': lambda x: x.mode().iloc[0] if len(x) > 0 else 'unknown',
            'drop_reason': lambda x: (x.notna()).sum(),
            'detecttime': lambda x: (x.max() - x.min()).total_seconds() if x.notna().any() else 0,
            'detect_date': 'nunique',
            'is_sybil': lambda x: x.any()
        }).rename(columns={
            'hash': 'total_transactions',
            'status': 'successful_transactions',
            'failurereason': 'failed_transactions',
            'toaddress': 'unique_recipients',
            'type': 'primary_tx_type',
            'region': 'primary_region',
            'drop_reason': 'dropped_transactions',
            'detecttime': 'activity_timespan_seconds',
            'detect_date': 'active_days',
            'is_sybil': 'is_sybil'
        })

        aggregated_features.append(tx_patterns)

        # Combine all aggregated features
        self.processed_df = pd.concat(aggregated_features, axis=1)

        # Calculate derived features
        self.processed_df['success_rate'] = (
            self.processed_df['successful_transactions'] /
            self.processed_df['total_transactions']
        ).fillna(0)

        self.processed_df['failure_rate'] = (
            self.processed_df['failed_transactions'] /
            self.processed_df['total_transactions']
        ).fillna(0)

        self.processed_df['unique_recipient_ratio'] = (
            self.processed_df['unique_recipients'] /
            self.processed_df['total_transactions']
        ).fillna(0)

        self.processed_df['drop_rate'] = (
            self.processed_df['dropped_transactions'] /
            self.processed_df['total_transactions']
        ).fillna(0)

        self.processed_df['tx_frequency'] = (
            self.processed_df['total_transactions'] /
            np.maximum(self.processed_df['active_days'], 1)
        )

        self.step_history.append("aggregation")
        self._show_dataset_overview(self.processed_df, "After Aggregation")

        print("✅ Feature aggregation completed!")
        return self.processed_df

    def preview_missing_values(self):
        """Preview missing values analysis"""
        if self.processed_df is None:
            print("❌ No processed data available. Run aggregation first.")
            return False

        print("\n🔍 PREVIEW: Missing Values Analysis")
        print("=" * 60)

        missing = self.processed_df.isnull().sum()
        missing_pct = (missing / len(self.processed_df) * 100).round(2)

        missing_df = pd.DataFrame({
            'Column': missing.index,
            'Missing_Count': missing.values,
            'Missing_Percentage': missing_pct.values
        })

        missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)

        if len(missing_df) == 0:
            print("✅ No missing values found!")
            return False

        print(f"📊 Columns with missing values ({len(missing_df)} total):")
        print(missing_df.to_string(index=False))

        # Show strategy for each column type
        numeric_cols = self.processed_df.select_dtypes(include=[np.number]).columns
        numeric_cols = numeric_cols.drop('is_sybil', errors='ignore')
        categorical_cols = self.processed_df.select_dtypes(include=['object']).columns

        print(f"\n🔧 Proposed Strategy:")
        print(f"  • Numeric columns ({len([c for c in missing_df['Column'] if c in numeric_cols])}): Fill with median")
        print(f"  • Categorical columns ({len([c for c in missing_df['Column'] if c in categorical_cols])}): Fill with mode")

        return input("\n❓ Proceed with missing value handling? (y/n): ").lower().strip() == 'y'

    def handle_missing_values(self, confirm=True):
        """Handle missing values with user confirmation"""
        if confirm and not self.preview_missing_values():
            print("❌ Missing value handling cancelled by user")
            return None

        print("\n🔄 Executing Missing Value Handling...")

        # Fill numeric columns with median
        numeric_columns = self.processed_df.select_dtypes(include=[np.number]).columns
        numeric_columns = numeric_columns.drop('is_sybil', errors='ignore')

        for col in numeric_columns:
            before_nulls = self.processed_df[col].isnull().sum()
            median_val = self.processed_df[col].median()
            self.processed_df[col] = self.processed_df[col].fillna(median_val)
            print(f"  • {col}: filled {before_nulls} nulls with median {median_val}")

        # Fill categorical columns with mode
        categorical_columns = self.processed_df.select_dtypes(include=['object']).columns
        for col in categorical_columns:
            before_nulls = self.processed_df[col].isnull().sum()
            mode_value = self.processed_df[col].mode().iloc[0] if len(self.processed_df[col].mode()) > 0 else 'unknown'
            self.processed_df[col] = self.processed_df[col].fillna(mode_value)
            print(f"  • {col}: filled {before_nulls} nulls with mode '{mode_value}'")

        self.step_history.append("missing_values")
        remaining_nulls = self.processed_df.isnull().sum().sum()
        print(f"\n✅ Missing value handling completed! Remaining nulls: {remaining_nulls}")

        return self.processed_df

    def preview_categorical_encoding(self):
        """Preview categorical encoding"""
        if self.processed_df is None:
            print("❌ No processed data available.")
            return False

        print("\n🔍 PREVIEW: Categorical Encoding Analysis")
        print("=" * 60)

        categorical_columns = self.processed_df.select_dtypes(include=['object']).columns
        categorical_columns = [col for col in categorical_columns if col != 'is_sybil']

        if len(categorical_columns) == 0:
            print("✅ No categorical columns to encode!")
            return False

        print(f"📊 Categorical columns to encode ({len(categorical_columns)}):")

        for col in categorical_columns:
            unique_values = self.processed_df[col].nunique()
            top_values = self.processed_df[col].value_counts().head(5)
            print(f"\n  • {col}:")
            print(f"    - Unique values: {unique_values}")
            print(f"    - Top 5 values:")
            for val, count in top_values.items():
                pct = count / len(self.processed_df) * 100
                print(f"      {val}: {count} ({pct:.1f}%)")

        print(f"\n🔧 Encoding Strategy: Label Encoding (0, 1, 2, ...)")

        return input("\n❓ Proceed with categorical encoding? (y/n): ").lower().strip() == 'y'

    def encode_categorical_features(self, confirm=True):
        """Encode categorical features"""
        if confirm and not self.preview_categorical_encoding():
            print("❌ Categorical encoding cancelled by user")
            return None

        print("\n🔄 Executing Categorical Encoding...")

        categorical_columns = self.processed_df.select_dtypes(include=['object']).columns
        categorical_columns = [col for col in categorical_columns if col != 'is_sybil']

        for col in categorical_columns:
            le = LabelEncoder()
            before_dtype = self.processed_df[col].dtype
            unique_before = self.processed_df[col].nunique()

            self.processed_df[col] = le.fit_transform(self.processed_df[col].astype(str))
            self.label_encoders[col] = le

            print(f"  • {col}: {before_dtype} -> numeric ({unique_before} categories)")
            print(f"    Mapping: {dict(zip(le.classes_[:3], range(3)))}{'...' if len(le.classes_) > 3 else ''}")

        self.step_history.append("encoding")
        print("✅ Categorical encoding completed!")

        return self.processed_df

    def preview_low_variance_removal(self, threshold=0.01):
        """Preview low variance feature removal"""
        if self.processed_df is None:
            print("❌ No processed data available.")
            return False, []

        print(f"\n🔍 PREVIEW: Low Variance Feature Removal (threshold={threshold})")
        print("=" * 60)

        feature_cols = self.processed_df.drop('is_sybil', axis=1, errors='ignore').select_dtypes(include=[np.number])
        variances = feature_cols.var().sort_values()

        low_var_features = variances[variances < threshold].index.tolist()

        if len(low_var_features) == 0:
            print("✅ No low variance features found!")
            return False, []

        print(f"📊 Features with variance < {threshold} ({len(low_var_features)} total):")
        for feature in low_var_features:
            var_val = variances[feature]
            unique_vals = self.processed_df[feature].nunique()
            print(f"  • {feature}: variance={var_val:.6f}, unique_values={unique_vals}")

            # Show value distribution for very low variance features
            if var_val < 0.001:
                top_values = self.processed_df[feature].value_counts().head(3)
                print(f"    Top values: {dict(top_values)}")

        print(f"\n📈 Features to keep ({len(feature_cols) - len(low_var_features)}):")
        keep_features = variances[variances >= threshold].sort_values(ascending=False).head(10)
        for feature, var_val in keep_features.items():
            print(f"  • {feature}: variance={var_val:.6f}")

        if len(low_var_features) > 0:
            proceed = input(f"\n❓ Remove {len(low_var_features)} low variance features? (y/n): ").lower().strip() == 'y'
            return proceed, low_var_features

        return False, []

    def remove_low_variance_features(self, threshold=0.01, confirm=True):
        """Remove features with low variance"""
        proceed, low_var_features = self.preview_low_variance_removal(threshold) if confirm else (True, [])

        if not proceed and confirm:
            print("❌ Low variance removal cancelled by user")
            return None

        if not confirm:  # Recompute if not confirming
            feature_cols = self.processed_df.drop('is_sybil', axis=1, errors='ignore').select_dtypes(include=[np.number])
            variances = feature_cols.var()
            low_var_features = variances[variances < threshold].index.tolist()

        if low_var_features:
            print(f"\n🔄 Removing {len(low_var_features)} low variance features...")
            self.processed_df = self.processed_df.drop(columns=low_var_features)
            for feature in low_var_features:
                print(f"  🗑️ Removed: {feature}")
            self.step_history.append("low_variance_removal")

        print("✅ Low variance removal completed!")
        return self.processed_df

    def preview_feature_selection(self, k=20):
        """Preview feature selection"""
        if self.processed_df is None or 'is_sybil' not in self.processed_df.columns:
            print("❌ No processed data with target available.")
            return False, []

        print(f"\n🔍 PREVIEW: Feature Selection (top {k} features)")
        print("=" * 60)

        X = self.processed_df.drop('is_sybil', axis=1)
        y = self.processed_df['is_sybil']

        print(f"📊 Current dataset:")
        print(f"  • Total features: {X.shape[1]}")
        print(f"  • Samples: {X.shape[0]}")
        print(f"  • Target distribution: {y.value_counts().to_dict()}")

        # Calculate feature scores
        print(f"\n🧮 Computing mutual information scores...")
        selector = SelectKBest(score_func=mutual_info_classif, k=min(k, X.shape[1]))
        selector.fit(X, y)

        # Feature importance scores
        feature_scores = pd.DataFrame({
            'feature': X.columns,
            'score': selector.scores_,
            'selected': selector.get_support()
        }).sort_values('score', ascending=False)

        print(f"\n🏆 Top 15 features by mutual information score:")
        print(feature_scores.head(15)[['feature', 'score']].to_string(index=False))

        if k < X.shape[1]:
            print(f"\n🔄 Features to be selected ({k}):")
            selected_features = feature_scores[feature_scores['selected']]['feature'].tolist()
            for i, feature in enumerate(selected_features, 1):
                score = feature_scores[feature_scores['feature'] == feature]['score'].iloc[0]
                print(f"  {i:2d}. {feature} (score: {score:.4f})")

            print(f"\n🗑️ Features to be removed ({X.shape[1] - k}):")
            removed_features = feature_scores[~feature_scores['selected']]['feature'].tolist()
            for feature in removed_features[:10]:  # Show first 10
                score = feature_scores[feature_scores['feature'] == feature]['score'].iloc[0]
                print(f"  • {feature} (score: {score:.4f})")
            if len(removed_features) > 10:
                print(f"  ... and {len(removed_features) - 10} more")

        proceed = input(f"\n❓ Proceed with selecting top {k} features? (y/n): ").lower().strip() == 'y'
        return proceed, feature_scores

    def feature_selection(self, k=20, confirm=True):
        """Select top k features based on mutual information"""
        if confirm:
            proceed, feature_scores = self.preview_feature_selection(k)
            if not proceed:
                print("❌ Feature selection cancelled by user")
                return None
        else:
            # Recompute for non-interactive mode
            X = self.processed_df.drop('is_sybil', axis=1)
            y = self.processed_df['is_sybil']
            selector = SelectKBest(score_func=mutual_info_classif, k=min(k, X.shape[1]))
            selector.fit(X, y)
            feature_scores = pd.DataFrame({
                'feature': X.columns,
                'score': selector.scores_,
                'selected': selector.get_support()
            }).sort_values('score', ascending=False)

        print(f"\n🔄 Executing feature selection...")

        X = self.processed_df.drop('is_sybil', axis=1)
        y = self.processed_df['is_sybil']

        selector = SelectKBest(score_func=mutual_info_classif, k=min(k, X.shape[1]))
        X_selected = selector.fit_transform(X, y)

        selected_features = X.columns[selector.get_support()].tolist()

        # Update processed dataframe
        self.processed_df = pd.concat([
            pd.DataFrame(X_selected, columns=selected_features, index=X.index),
            y
        ], axis=1)

        self.feature_importance = feature_scores
        self.step_history.append("feature_selection")

        print(f"✅ Feature selection completed! Selected {len(selected_features)} features")

        return selected_features

    def show_current_state(self):
        """Show current state of the pipeline"""
        print(f"\n📋 Pipeline Current State")
        print("=" * 40)
        print(f"Steps completed: {', '.join(self.step_history) if self.step_history else 'None'}")

        if self.processed_df is not None:
            self._show_dataset_overview(self.processed_df, "Current Processed Dataset")
        else:
            print("No processed data available yet.")

    def plot_feature_analysis(self):
        """Create visualizations for feature analysis"""
        if self.processed_df is None:
            print("❌ No processed data available for plotting")
            return

        print("📊 Creating feature analysis plots...")

        fig, axes = plt.subplots(2, 2, figsize=(15, 12))

        # 1. Target distribution
        if 'is_sybil' in self.processed_df.columns:
            target_counts = self.processed_df['is_sybil'].value_counts()
            axes[0, 0].pie(target_counts.values, labels=['Non-Sybil', 'Sybil'], autopct='%1.2f%%')
            axes[0, 0].set_title('Target Distribution (is_sybil)')

        # 2. Feature importance (if available)
        if self.feature_importance is not None:
            top_features = self.feature_importance.head(15)
            y_pos = range(len(top_features))
            axes[0, 1].barh(y_pos, top_features['score'])
            axes[0, 1].set_yticks(y_pos)
            axes[0, 1].set_yticklabels([f[:20] + '...' if len(f) > 20 else f for f in top_features['feature']])
            axes[0, 1].set_title('Top 15 Feature Importance Scores')
            axes[0, 1].set_xlabel('Mutual Information Score')

        # 3. Correlation heatmap of top features
        if self.feature_importance is not None:
            top_feature_names = self.feature_importance.head(8)['feature'].tolist()
            corr_features = [f for f in top_feature_names if f in self.processed_df.columns]
            if corr_features:
                corr_matrix = self.processed_df[corr_features].corr()
                sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
                          square=True, ax=axes[1, 0], fmt='.2f')
                axes[1, 0].set_title('Correlation Matrix (Top Features)')

        # 4. Feature distribution comparison
        if 'is_sybil' in self.processed_df.columns:
            key_features = ['total_transactions', 'unique_recipients', 'success_rate', 'tx_frequency']
            available_features = [f for f in key_features if f in self.processed_df.columns]

            if available_features:
                feature = available_features[0]
                sybil_data = self.processed_df[self.processed_df['is_sybil'] == True][feature]
                non_sybil_data = self.processed_df[self.processed_df['is_sybil'] == False][feature]

                # Sample data if too large for visualization
                if len(non_sybil_data) > 1000:
                    non_sybil_data = non_sybil_data.sample(1000, random_state=42)

                axes[1, 1].boxplot([non_sybil_data, sybil_data], labels=['Non-Sybil', 'Sybil'])
                axes[1, 1].set_title(f'Distribution: {feature}')
                axes[1, 1].set_ylabel('Value')

        plt.tight_layout()
        plt.show()

    def prepare_for_modeling(self, test_size=0.2, random_state=42):
        """Final preparation for machine learning model"""
        if self.processed_df is None or 'is_sybil' not in self.processed_df.columns:
            print("❌ No processed data with target available")
            return None

        print("\n🔄 Preparing data for modeling...")

        X = self.processed_df.drop('is_sybil', axis=1)
        y = self.processed_df['is_sybil']

        # Scale features
        X_scaled = pd.DataFrame(
            self.scaler.fit_transform(X),
            columns=X.columns,
            index=X.index
        )

        # Train-test split
        X_train, X_test, y_train, y_test = train_test_split(
            X_scaled, y, test_size=test_size, random_state=random_state,
            stratify=y
        )

        print(f"✅ Data prepared for modeling:")
        print(f"   Training set: {X_train.shape}")
        print(f"   Test set: {X_test.shape}")
        print(f"   Sybil ratio in train: {y_train.sum()/len(y_train):.3f}")
        print(f"   Sybil ratio in test: {y_test.sum()/len(y_test):.3f}")

        return X_train, X_test, y_train, y_test

    def run_baseline_model(self, X_train, X_test, y_train, y_test):
        """Run a baseline Random Forest model for validation"""
        print("🤖 Training baseline Random Forest model...")

        rf = RandomForestClassifier(
            n_estimators=100,
            class_weight='balanced',
            random_state=42,
            n_jobs=-1
        )

        rf.fit(X_train, y_train)

        y_pred = rf.predict(X_test)
        y_pred_proba = rf.predict_proba(X_test)

        print("\n📊 Baseline Model Performance:")
        print(classification_report(y_test, y_pred))

        model_importance = pd.DataFrame({
            'feature': X_train.columns,
            'importance': rf.feature_importances_
        }).sort_values('importance', ascending=False)

        print("\nTop 10 Most Important Features (Random Forest):")
        print(model_importance.head(10).to_string(index=False))

        return rf, y_pred, y_pred_proba, model_importance

    def run_full_pipeline(self, k_features=20):
        """Run the complete pipeline non-interactively"""
        print("🚀 Running Full Pipeline (Non-Interactive Mode)")
        print("=" * 60)

        self.aggregate_features(confirm=False)
        self.handle_missing_values(confirm=False)
        self.encode_categorical_features(confirm=False)
        self.remove_low_variance_features(confirm=False)
        self.feature_selection(k=k_features, confirm=False)
        self.plot_feature_analysis()
```

```python
import pandas as pd

def update_pending_times_and_clean(df: pd.DataFrame) -> pd.DataFrame:
    print("🔁 Updating pending times and removing raw pending rows...")

    # Make sure detecttime is datetime
    df['detecttime'] = pd.to_datetime(df['detecttime'], errors='coerce')

    # Split pending and confirmed sets
    pending_df = df[df['status'] == 'pending']
    confirmed_df = df[df['status'] == 'confirmed']

    # Keep only hashes that are present in both
    common_hashes = set(pending_df['hash']).intersection(set(confirmed_df['hash']))

    # Only process those
    pending_df = pending_df[pending_df['hash'].isin(common_hashes)]
    confirmed_df = confirmed_df[confirmed_df['hash'].isin(common_hashes)]

    # Map pending times to confirmed df
    pending_times = pending_df[['hash', 'detecttime']].rename(columns={'detecttime': 'pending_detecttime'})
    confirmed_df = confirmed_df.merge(pending_times, on='hash', how='left')

    # Compute time diff in seconds
    confirmed_df['computed_time_pending'] = (
        confirmed_df['detecttime'] - confirmed_df['pending_detecttime']
    ).dt.total_seconds()

    # Update time_pending column
    confirmed_df['time_pending'] = confirmed_df.apply(
        lambda row: row['computed_time_pending']
        if pd.isna(row['time_pending']) else max(row['time_pending'], row['computed_time_pending']),
        axis=1
    )

    # Drop helper columns
    confirmed_df = confirmed_df.drop(columns=['pending_detecttime', 'computed_time_pending'])

    # Remove original pending rows
    df_cleaned = df[df['status'] != 'pending']

    # Replace old confirmed rows with updated ones
    df_cleaned = df_cleaned[~df_cleaned['hash'].isin(common_hashes)]
    df_cleaned = pd.concat([df_cleaned, confirmed_df], ignore_index=True)

    print("✅ time_pending updated and pending rows removed.")
    return df_cleaned
```
