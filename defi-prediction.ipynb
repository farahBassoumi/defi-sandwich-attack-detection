```python


# Optional: convert block_time to datetime if needed
df['block_time'] = pd.to_datetime(df['block_time'])

# Set style
sns.set(style="whitegrid")

# Plot: Total USD traded over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="block_time", y="amount_usd")
plt.title("Uniswap V2 Trades on 2024-01-13 (USD Volume Over Time)")
plt.xlabel("Time")
plt.ylabel("Amount (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

suspected_sandwiches = []

# Group by block
for block_time, block_df in df.groupby("block_time"):
    trades = block_df.reset_index(drop=True)

    for i in range(len(trades) - 2):
        t1, t2, t3 = trades.iloc[i], trades.iloc[i+1], trades.iloc[i+2]

        # Check addresses
        if (
            t1["tx_from"] == t3["tx_from"] and  # same attacker
            t1["tx_from"] != t2["tx_from"]      # victim is different
        ):
            suspected_sandwiches.append({
                "block_time": block_time,
                "attacker": t1["tx_from"],
                "victim": t2["tx_from"],
                "token_pair": t1["token _pair"],
                "amount_usd": t2["amount_usd"],
                "t1_tx": t1["tx_hash"],
                "t2_tx": t2["tx_hash"],
                "t3_tx": t3["tx_hash"]
            })

# Convert to DataFrame
suspect_df = pd.DataFrame(suspected_sandwiches)
print(suspect_df.head())
```
